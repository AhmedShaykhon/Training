### Reading an image

### Syntax: cv2.imread(path, flag)

**Parameters:**

1. path: A string representing the path of the image to be read.
2. flag: It specifies the way in which image should be read. Itâ€™s default value is cv2.IMREAD_COLOR

3. Return Value: This method returns an image that is loaded from the specified file.


- cv2.IMREAD_COLOR: It specifies to load a color image. Any transparency of image will be neglected. It is the default flag. Alternatively, we can pass integer value 1 for this flag.
- cv2.IMREAD_GRAYSCALE: It specifies to load an image in grayscale mode. Alternatively, we can pass integer value 0 for this flag.
- cv2.IMREAD_UNCHANGED: It specifies to load an image as such including alpha channel. Alternatively, we can pass integer value -1 for this flag.

### Dispay Image Function 
- imshow(Label, image)

```python
def ImageShow(image):
        cv2.imshow('imageLabel', image) 
        k = cv2.waitKey(0) & 0xFF

# wait for ESC key to exit 
        if k == 27: 
            cv2.destroyAllWindows()
```

```python
def DisplayImage(image,title,cols=1,axis ='on'):
  
    import matplotlib.pyplot as plt
    
    image_no=len(image)+1
    postion=0

    plt.figure(figsize=(12, 12))
    for i in range(1,image_no):
        postion+=1
        plt.subplot(1,cols,postion)
        plt.imshow(image[i-1],cmap = 'gray')
        plt.title(title[i-1])
        if(axis == 'off'):
            plt.axis('off')
        
       
        if ( i%cols==0):
            plt.show(),plt.figure(figsize=(12, 12))
            postion=0
```

```python
# Importing the OpenCV library 
import cv2 
# Reading the image using imread() function 
image = cv2.imread('img/road.jpg') 

# Extracting the height and width of an image 
h, w = image.shape[:2] 
# (1603, 2400, 3) 3 is channel number
print(image.shape)
# Displaying the height and width 
print("Height = {}, Width = {}".format(h, w)) 
ImageShow(image)
title=[]
title.append("Road Image")
imageN=[]
imageN.append(image)
DisplayImage(imageN,title,1)
```

### Extracting the RGB values of a pixel

```python
# Extracting RGB values. 
# Here we have randomly chosen a pixel 
# by passing in 100, 100 for height and width. 
(B, G, R) = image[100, 100] 

# Displaying the pixel values 
print("R = {}, G = {}, B = {}".format(R, G, B)) 

# We can also pass the channel to extract 
# the value for a specific channel 
B = image[100, 100, 0] 
print("B = {}".format(B)) 
```

### Extracting the Region of Interest (ROI)

```python
# We will calculate the region of interest 
# by slicing the pixels of the image 
roi = image[100 : 500, 200 : 700] 

ImageShow(roi)

title.append("Region of Interest ")
imageN.append(roi)
DisplayImage(imageN,title,2)
```

### Resizing the Image

```python
# resize() function takes 2 parameters, 
# the image and the dimensions 
resize = cv2.resize(image, (800, 800)) 
```

```python
# Calculating the ratio 
ratio = 800 / w   
print (ratio)# what is the aim from this ratio
# Creating a tuple containing width and height 
dim = (800, int(h * ratio)) 
print(dim)
# Resizing the image 
resize_aspect = cv2.resize(image, dim) 
#print(resize_aspect)


ImageShow(resize_aspect)

title=[]
title.append("Road Image")
title.append("resize_aspect")
imageN=[]
imageN.append(image)
imageN.append(resize_aspect)
DisplayImage(imageN,title,2)
```

### Rotating the Image

```python
# Calculating the center of the image 
center = (w // 2, h // 2) 

# Generating a rotation matrix
# getRotationMatrix2D(center,Angle,Scale)

matrix = cv2.getRotationMatrix2D(center, -45, 1.0) 

# Performing the affine transformation 
rotated = cv2.warpAffine(image, matrix, (w, h)) 

#print(rotated)

ImageShow(rotated)

title=[]
title.append("Road Image")
title.append("rotated Image")
imageN=[]
imageN.append(image)
imageN.append(rotated)
DisplayImage(imageN,title,2)
```

### Drawing a Rectangle

```python
# We are copying the original image, 
# as it is an in-place operation. 
output = image.copy() 

# Using the rectangle() function to create a rectangle. 
rectangle = cv2.rectangle(output, (1500, 900), 
						(600, 400), (255, 0, 0), 2) 

#print(rectangle)



ImageShow(rectangle)


title=[]
title.append("Road Image")
title.append("rectangle")
imageN=[]
imageN.append(image)
imageN.append(rectangle)
DisplayImage(imageN,title,2)
```

### Displaying text

```python
# Copying the original image 
output = image.copy() 

# Adding the text using putText() function 
text = cv2.putText(output, 'OpenCV Demo', (500, 550), 
				cv2.FONT_HERSHEY_SIMPLEX, 4, (255, 0, 0), 2) 

#print(text)

ImageShow(text)

title=[]
title.append("text")
title.append("rectangle")
imageN=[]
imageN.append(rectangle)
imageN.append(text)

DisplayImage(imageN,title,2)
```
