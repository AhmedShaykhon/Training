### Addition of Image:

### Syntax: cv2.addWeighted(img1, wt1, img2, wt2, gammaValue)

**Parameters:**

1. img1: First Input Image array(Single-channel, 8-bit or floating-point)
2. wt1: Weight of the first input image elements to be applied to the final image
3. img2: Second Input Image array(Single-channel, 8-bit or floating-point)
4. wt2: Weight of the second input image elements to be applied to the final image
5. gammaValue: Measurement of light

```python
def DisplayImage(image,title,cols=1):
  
    import matplotlib.pyplot as plt
    
    image_no=len(image)+1
    postion=0

    plt.figure(figsize=(12, 12))
    for i in range(1,image_no):
        postion+=1
        plt.subplot(1,cols,postion)
        plt.imshow(image[i-1],cmap = 'gray')
        plt.title(title[i-1])
        plt.axis('off')
       
        if ( i%cols==0):
            plt.show(),plt.figure(figsize=(8, 8))
            postion=0
```

```python
# Python code to reading an image using OpenCV 
import cv2 
import numpy as np 
import matplotlib.pyplot as plt 
# You can give path to the 
# image as first argument 
img1 = cv2.imread('img/cc-2.png', 0) 
img2 = cv2.imread('img/messigray.png', 0) 
print(type(img1))
img3 = cv2.add(img1, img2)

title=[]
title.append("Dog")
title.append("Dog")
title.append("Dog")
imageN=[]
imageN.append(img1)
imageN.append(img2)
imageN.append(img3)
DisplayImage(imageN,title,3)


cv2.imshow('image', img3) 
k = cv2.waitKey(0) & 0xFF

# wait for ESC key to exit 
if k == 27: 
	cv2.destroyAllWindows() 
```

```python
# Python programe to illustrate  
# arithmetic operation of 
# addition of two images 
    
# organizing imports  
import cv2  
import numpy as np  
    
# path to input images are specified and   
# images are loaded with imread command  
image1 = cv2.imread('img/image4.1.jpg')  
image2 = cv2.imread('img/image4.2.jpg') 
  
# cv2.addWeighted is applied over the 
# image inputs with applied parameters 
weightedSum = cv2.addWeighted(image1, 0.5, image2, 0.4, 0) 


title=[]
title.append("image1")
title.append("image2")
title.append("weightedSum")
imageN=[]
imageN.append(image1)
imageN.append(image2)
imageN.append(weightedSum)
DisplayImage(imageN,title,3)

# the window showing output image 
# with the weighted sum  
cv2.imshow('Weighted Image', weightedSum) 
  
# De-allocate any associated memory usage   
if cv2.waitKey(0) & 0xff == 27:  
    cv2.destroyAllWindows()  
```

### Subtraction of Image:

### Syntax:  cv2.subtract(src1, src2)

```python
# Python programe to illustrate 
# arithmetic operation of 
# subtraction of pixels of two images 

# organizing imports 
import cv2 
import numpy as np 
	
# path to input images are specified and 
# images are loaded with imread command 
image1 = cv2.imread('img/image4.3.jpg') 
image2 = cv2.imread('img/image4.4.jpg') 

# cv2.subtract is applied over the 
# image inputs with applied parameters 
sub = cv2.subtract(image1, image2) 

title=[]
title.append("image1")
title.append("image2")
title.append("sub")
imageN=[]
imageN.append(image1)
imageN.append(image2)
imageN.append(sub)
DisplayImage(imageN,title,3)


# the window showing output image 
# with the subtracted image 
cv2.imshow('Subtracted Image', sub) 

# De-allocate any associated memory usage 
if cv2.waitKey(0) & 0xff == 27: 
	cv2.destroyAllWindows() 
```
