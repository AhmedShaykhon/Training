# 4-Question
Write a Pandas program to find the details of the diamonds where length>5, width>5 and depth>5. (x,y,z)

```python
df_diamonds.loc[(df_diamonds['x']>5)| (df_diamonds['y']>5)|(df_diamonds['z']>5), ['x','y','z']]
```
